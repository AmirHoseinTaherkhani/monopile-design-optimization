```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
%matplotlib inline
```

```python
H05 = pd.read_csv('H05.csv').drop('Unnamed: 0', axis = 1)
H05['Train_Loss'] = H05['Train_Loss'] * 100 
H05['Val_Loss'] = H05['Val_Loss'] * 100 


H1 = pd.read_csv('H1.csv').drop('Unnamed: 0', axis = 1)
H1['Train_Loss'] = H1['Train_Loss'] * 100 
H1['Val_Loss'] = H1['Val_Loss'] * 100 
```

```python
def vis(epoch, train_loss, val_loss, output):
    
    plt.figure(figsize = (6,4), facecolor = (1, 1, 1))
    
    plt.plot(epoch, train_loss, c='tab:blue', label = 'Training loss')
    plt.plot(epoch, val_loss,'--', c='tab:red', label = 'Validation loss')
    
#     plt.ylim((0,600))
    plt.ylim((0.0,50.0))
    
    plt.rc('xtick', labelsize=14) 
    plt.rc('ytick', labelsize=14)
    plt.rc('font', **{'family' : 'Arial','weight' : 'normal'})
    
    plt.xlabel('Epoch', fontname="Arial", fontsize=16)
    plt.ylabel('MAPE (%)', fontname="Arial", fontsize=16)
    
    plt.legend(prop={'size': 14})
    plt.tight_layout()
    plt.savefig(output+'.png', dpi = 300)
```

```python
vis(H05['Epoch'], H05['Train_Loss'], H05['Val_Loss'], output = 'H0.5')
```

```python
vis(H1['Epoch'], H1['Train_Loss'], H1['Val_Loss'], output = 'H1')
```
