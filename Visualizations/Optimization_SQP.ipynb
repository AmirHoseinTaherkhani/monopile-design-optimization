```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rc('font',family='Arial')
```

```python
data = pd.read_csv('data.csv')
```

```python
from matplotlib.ticker import MaxNLocator
import matplotlib as mpl


fig, ax1 = plt.subplots(figsize=(8, 5))
ax2 = ax1.twinx()

line1 = ax1.plot(np.arange(len(data)), data['weight'], marker='v', color='royalblue', lw=2, label = 'Weight History')
# line15 = ax1.scatter(np.arange(len(data)), data['weight'],marker='v', color='royalblue')

line2 = ax1.plot(np.arange(len(data)), data['initial_weight'][0]*np.ones((len(data)),), linestyle = 'dashdot', lw=1, color = 'royalblue', label = 'Initial Weight')

line3 = ax2.plot(np.arange(len(data)), data['capacity1'], marker = 'o', color='deeppink', lw=2, label = 'Capacity History')
# line35 = ax2.scatter(np.arange(len(data)), data['capacity1'], color='deeppink')

line4 = ax2.plot(np.arange(len(data)), data['target_capacity1'][0]*np.ones((len(data)),), linestyle = 'dotted', lw=1, color = 'deeppink', label = 'Desired Capacity')

ax1.set_xlabel('Iteration', fontname="Arial", fontsize=16)
ax1.xaxis.set_major_locator(MaxNLocator(integer=True))
ax1.set_xlim(0,17)

ax1.set_ylabel(r'Weight (kg) = $ρπLBt_{w}$', fontname="Arial", fontsize=16)
ax1.set_ylim(0,800000)
ax1.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))


ax2.set_ylabel(r'$H_{1}$ (kN)', fontname="Arial", fontsize=16)
ax2.set_ylim(0,22500)
ax2.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))


lns = line1+line2+line3+line4
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc = 'lower right', fontsize=13)
plt.grid(alpha = 0.2)
plt.savefig('Optimization2.jpg', dpi = 300)
```
